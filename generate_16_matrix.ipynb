```python
import numpy as np
from tqdm import tqdm
import os
```

```python
dir_name = 'matrix_16'
matrix_size=16
if not os.path.isdir(dir_name):
    os.mkdir(dir_name)

base_matrix = np.random.randint(5, size=(16, 16))-2
inv_base_matrix = np.linalg.inv(base_matrix)
```

```python
np.save(dir_name+'/'+'base_matrix', base_matrix)
np.save(dir_name+'/'+'inv_base_matrix', inv_base_matrix)
```

```python
epsilon=0.01
temp = np.random.rand(matrix_size,matrix_size).astype(np.float64)
temp = temp*2-1
cur_matrix = base_matrix + epsilon*temp
temp_inv = np.linalg.inv(cur_matrix)
print(temp)
print(cur_matrix)
print(temp_inv)
```
